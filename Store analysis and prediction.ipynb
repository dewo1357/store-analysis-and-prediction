```python
from sklearn.cluster import KMeans

x = df[['Store_Area','Items_Available']]

clusters = []

for i in range(1,11):
    km=KMeans(n_clusters=i).fit(x)
    clusters.append(km.inertia_)
```

```python
fig, ax = plt.subplots(figsize=(10,3))
sns.lineplot(x=list(range(1,11)), y=clusters,ax=ax)
plt.show()
km.score(x)
```

```python
km5 = KMeans(n_clusters=3).fit(x)
x['labels'] = km5.labels_

plt.figure()
sns.scatterplot(x=x['Items_Available'],y=x['Store_Area'],hue=x['labels'],data=x)
```

```python
df['Category_Yards'] = x['labels']
df['Category_Yards'].replace({2 : 'Low Yards', 1 : 'Standard Yards',0 : 'High Yards' },inplace=True)
df
```

```python
df = df[['Store_Area','Items_Available','Daily_Customer_Count','Store_Sales','Category_Yards']]
df.info()
```

```python
X = df.drop(['Store_Sales','Category_Yards'],axis=1)
y = df['Store_Sales']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.30, random_state=42)
```

```python
column = ['Store_Area','Items_Available','Daily_Customer_Count']
```

```python
from sklearn.preprocessing import MinMaxScaler
mmscaler = MinMaxScaler(feature_range=(0,1))
X_train[column]=mmscaler.fit_transform(X_train[column])
X_test[column]=mmscaler.fit_transform(X_test[column])
X_train = pd.DataFrame(X_train)
X_test = pd.DataFrame(X_test)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model = model.fit(X_train,y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn import metrics

print('R squared :',round(metrics.r2_score(y_test,y_pred),6))
print('Mean Absolut Error :',round(metrics.mean_absolute_error(y_test,y_pred),3))
print('Mean Squared Error :',round(metrics.mean_squared_error(y_test,y_pred),3))
print('Root Mean Squad Error :',round(np.sqrt(metrics.mean_squared_error(y_test,y_pred)),3))

def MAPE(y_test,y_pred):
    y_test, y_pred = (np.array(y_test),np.array(y_pred))
    return (np.mean(np.abs(y_test - y_pred) / y_test)) * 100

result = MAPE(y_test,y_pred)
print('Mean Absolute Square Error Percentage :', round(result), '%')
```

```python
result = pd.DataFrame({'Store_Sales_A': y_test, 'Store_Sales_B' : y_pred})

data = df.merge(result, left_index=True, right_index=True)
data.drop('Category_Yards',axis=1,inplace=True)
data.drop('Store_Sales',axis=1,inplace=True)
data.rename(columns=({
    'Store_Sales_A' : 'Actual_Sales',
    'Store_Sales_B' : 'prediction'
}),inplace=True)
data.sample(10)
```
